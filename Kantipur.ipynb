```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
main_url = 'https://ekantipur.com/'
main_page = requests.get(main_url)
main_page
soup = BeautifulSoup(main_page.content, 'html')
```

```python
headers = soup.find_all('div', class_ = 'bottom-nav')
headers
```

```python
headers = soup.find_all('div', class_='bottom-nav')

for title in headers:
    text = title.get_text("\n").strip()   

    text = text.replace("Subscribe To Kantipur", "")
    text = text.replace("Fact Checker", "")

    print(text)
```
